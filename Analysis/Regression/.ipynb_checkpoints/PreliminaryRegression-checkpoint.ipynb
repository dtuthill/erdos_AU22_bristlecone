Because EDA over this long list of variables is fairly laborious, one way to decide whether we should be regressing on some scaled total of accidents or the log of that total is to look at the probability density function. Here, you will find that that distribution fairly closely follows that of a Landau distribution:

```python
from scipy.stats import moyal
from scipy.optimize import curve_fit

def func(x, loc, scale):
    return moyal.pdf(x, loc, scale)
```

```python
plt.figure(figsize=(10,6))

hist, bins = np.histogram(data_raw.loc[data_raw.total > 5].total_dnorm_bvnorm, bins=40, range=[0,40], density=True)
bins = bins[1:]-0.5

par, cov = curve_fit(func, bins, hist)

plt.plot(bins, hist, 'o',
            label='bins')
plt.plot(bins, moyal.pdf(bins, *par), 'r-',
            label='fit: loc=%5.3f, scale=%5.3f' % tuple(par))

plt.xlabel('total_dnorm_bvnorm', fontsize=16)
plt.ylabel('probability density', fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(fontsize=16)

plt.show()
```

Here, scaling the data is important because of the number of binary variables, which are naturally outweighed by the continuous ones without scaling.

```python
## import StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
## Make a scaler object
scaler = StandardScaler()

## fit the scaler
data_raw[['aadf_Pedal_cycles_scaled', 'aadf_All_motor_vehicles_scaled', 'distance_to_cp_scaled']] = scaler.fit_transform(data_raw[['aadf_Pedal_cycles', 'aadf_All_motor_vehicles', 'distance_to_cp']])
```

```python
reg_features = ['aadf_Pedal_cycles_scaled', 'aadf_All_motor_vehicles_scaled',
                    'aadf_CLT_CARR', 'aadf_CLT_SEGREG', 'aadf_CLT_STEPP', 
                    'aadf_CLT_PARSEG', 'aadf_CLT_SHARED', 'aadf_CLT_MANDAT',
                    'aadf_CLT_ADVIS', 'aadf_CLT_PRIORI', 'aadf_CLT_CONTRA', 
                    'aadf_CLT_BIDIRE', 'aadf_CLT_CBYPAS', 'aadf_CLT_BBYPAS', 
                    'aadf_CLT_PARKR', 'aadf_CLT_WATERR', 'aadf_CLT_PTIME', 
                    'distance_to_cp_scaled', 'bikelane']

data = data_raw.copy()
```

```python
sns.pairplot(data[['total_dnorm_bvnorm', 'aadf_Pedal_cycles_scaled', 'aadf_All_motor_vehicles_scaled', 'distance_to_cp_scaled']])

plt.show()
```

It is fairly clear from this small amount of EDA on the continuous variables, that we should at least be taking the log of our normalized accident total.

```python
data['log_total_dnorm_bvnorm'] = np.log(data.total_dnorm_bvnorm)
sns.pairplot(data[['log_total_dnorm_bvnorm', 'aadf_Pedal_cycles_scaled', 'aadf_All_motor_vehicles_scaled', 'distance_to_cp_scaled']])

plt.show()
```

While there are clearly other data manipulations (some kind of polynomial/funcitonal/interaction term addition) that is necessary to get meaningful trends for visualization, but these are, unfortunately, beyond the scope of this project for time reasons.

Finally, make a train test split for preliminary regression to work on:

```python
## import train_test_split
from sklearn.model_selection import train_test_split
```

```python
## make the validation set
# Temporary fix
data.replace([np.inf, -np.inf], np.nan, inplace=True)
# Drop rows with NaN
data.dropna(inplace=True)

data_train, data_test = train_test_split(data,
                                             shuffle = True,
                                             random_state = 9483,
                                             test_size = .1)
```

One additional complication with this dataset is that stratifying by the binary variables could be helpful to make sure that our splits have an even distribution of different bike safety features.

##### 1. Ridge and Lasso Regression

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
alphas = [0.000001,.00001,.0001,.001,.01,.1,.25,.5,.75,1,5,10,25,50,100,1000,10000]

## make coefficient holder
lasso_coefs = np.zeros((len(alphas), len(reg_features)))
ridge_coefs = np.zeros((len(alphas), len(reg_features)))

## Loop through alphas
for i in range(len(alphas)):
    ## make lasso model
    lasso = Lasso(alpha=alphas[i], max_iter=10000000)
    ridge = Ridge(alpha=alphas[i], max_iter=10000000)
    
    ## fit model
    lasso.fit(data_train[reg_features], data_train.log_total_dnorm_bvnorm)
    ridge.fit(data_train[reg_features], data_train.log_total_dnorm_bvnorm)
    
    ## record coefs
    lasso_coefs[i,:] = lasso.coef_
    ridge_coefs[i,:] = ridge.coef_
    
```

```python
pd.DataFrame(lasso_coefs, index=alphas, columns=reg_features)
```

```python
pd.DataFrame(ridge_coefs, index=alphas, columns=reg_features)
```

It is fairly clear from the Lasso and Ridge regression that the `aadf_Pedal_cycles_scaled` dominates in importance, which is not surprising as one should expect there to be more accidents if there are more cyclists on the road. This correlation has been shown in EDA previously. Other bike safety variables are otherwise all of fairly equal importance.

##### 2. Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression(copy_X=True)

slr.fit(data_train[reg_features], data_train.log_total_dnorm_bvnorm)

residuals = data_test.log_total_dnorm_bvnorm - slr.predict(data_test[reg_features])
```

```python
residuals
```

##### 3. GridSearchCV

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
grid_cv = GridSearchCV(KNeighborsRegressor(), 
                         param_grid = {'n_neighbors':range(1,50),
                                          'weights':['uniform', 'distance']}, 
                         scoring = 'neg_mean_squared_error',
                         cv = KFold(5, shuffle=True, random_state = 30293))
```

```python
grid_cv.fit(data_train[reg_features], data_train.log_total_dnorm_bvnorm)
```

```python
grid_cv.best_params_
```

```python
grid_cv.best_estimator_.predict(data_test[reg_features])
```

```python
grid_cv.best_score_
```
